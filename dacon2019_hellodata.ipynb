```
import warnings
import pandas as pd
warnings.filterwarnings('ignore')

# Train 데이터를 불러옵니다.
train = pd.read_csv('drive/My Drive/filmthin/train.csv', sep=',')
# Test 데이터를 불러옵니다.
test = pd.read_csv('drive/My Drive/filmthin/test.csv', sep=',')
```

```
#독립변수와 종속변수를 분리합니다.
train_X = train.iloc[:,4:]  #Train 데이터의 독립변수
train_Y = train.iloc[:,0:4] #Train 데이터의 종속변수
test_X = test.iloc[:,1:]    #Test  데이터의 독립변수
```

```
#케라스를 통해 모델 생성을 시작합니다.
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(units=160, activation='relu', input_dim=226))
model.add(Dense(units=160, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=160, activation='relu'))
model.add(Dense(units=160, activation='relu'))
model.add(Dense(units=160, activation='relu'))
model.add(Dense(units=160, activation='relu'))
model.add(Dense(units=160, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=4, activation='linear'))
```

```
#모델을 컴파일합니다.
adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)
model.compile(loss='mae', optimizer=adam, metrics=['mae'])
```

```
#모델을 학습합니다.
model.fit(train_X, train_Y, epochs=300, batch_size= 3000, validation_split = 0.05)
```

```
#예측값을 생성합니다.
pred_test = model.predict(test_X)
```

```
#submission 파일을 생성합니다.
sample_sub = pd.read_csv('drive/My Drive/filmthin/sample_submission.csv', index_col=0)
submission = sample_sub+pred_test
submission.to_csv('./submission.csv')
```
